```python
import pandas as pd

movie=pd.read_csv("nlp2_data",encoding = 'utf_8_sig')
movie.head()
```

```python
#修改類型
for i in range(0,len(movie)):
    type=''
    if str(movie['類型'][i])=='nan':
        continue
    for k in movie['類型'][i]:
        if str(k) !='、':
            type=type+str(k)
        else:
            break
    movie['類型'][i]=type
movie.head()

```
