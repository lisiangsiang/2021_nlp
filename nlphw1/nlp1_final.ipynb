```python
import jieba
import jieba.analyse
import requests

r = requests.get('https://raw.githubusercontent.com/cjwu/cjwu.github.io/master/courses/nlp/hw1-dataset.txt')
r.encoding = 'utf-8'
doc = jieba.lcut(r.text)
count={}
for word in doc:
  if len(word)!=1:
    count[word] = count.get(word,0)+1
items = list(count.items())
items.sort(key=lambda x:x[1],reverse=True)
x=[]
y=[]
for n in range(100):
  word, count = items[n]
  x.append(word)
  y.append(count)
  print("{0:<10}{1:>5}".format(word,count))
allword=0 #所有詞的數量
for m in range(len(items)):
  word, count = items[m]
  allword=allword+count
print(allword)
```

```python
words = open('words.txt','w')
words.write(r.text)
words.close
f=open('words.txt')
lines=0
for line in f.readlines():
  lines=lines+1 #文件數量

print(lines)
```

```python
#fig#2
import numpy as np
import matplotlib.pyplot as plt

label=x
y_array=np.array(y)
fre_array=y_array/allword
plt.bar(x,fre_array , tick_label=label)
plt.show()
print('fig#2')
```

```python
#fig#1
import math

tfidf=[]
for i in range(100):
  idf=math.log(lines/y[i])
  tfidf.append(fre_array[i]*idf)
plt.bar(x,tfidf , tick_label=label)
plt.show()
print('fig#1')
```
